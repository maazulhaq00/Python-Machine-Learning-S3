```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from google.colab import drive
```

```python
car_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/datasets/carprices.csv')

car_df.head()
```

```python
for col in car_df.columns:
  if col != 'Sell Price($)':
    plt.title(f"Car Price vs {col}")
    plt.xlabel(f"{col}")
    plt.ylabel(f"Sell Price (dollars)")
    plt.scatter(car_df[col], car_df['Sell Price($)'])
    plt.show()
```

- axis=0 (or 'index'): Refers to the rows or index.
Operations are applied down each column (e.g., calculating a mean for each column).
- axis=1 (or 'columns'): Refers to the columns. Operations are applied across each row (e.g., calculating a mean for each row).

```python
# X = car_df[['Mileage', 'Age(yrs)']]
X = car_df.drop('Sell Price($)', axis=1)
y = car_df['Sell Price($)']

print(X.head())
print(y.head())
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=59)
```

```python
X_test
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```
